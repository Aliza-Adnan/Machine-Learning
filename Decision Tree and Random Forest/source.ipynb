```python
import pandas as pd
df = pd.read_csv ('adult.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
y=df[' Male']
g=df.drop([' Male'],axis=1)
g.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(g,y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_leaf_nodes=15,random_state=40)
clf=clf.fit(X_train[[' 40',' 77516','39']],y_train)
clf
```

```python
pr=clf.predict(X_test[[' 40',' 77516','39']])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pr)
```
